# 3. Data Preparation and Preprocessing
This include steps such as :removing stopwords,  tokenizing the text,  and performing any other necessary transformations. Additionally, splitting of the dataset into training and testing sets before modeling.

### Removing Stop Words and placeholders like {{USERNAME}} and {{URL}} and also selecting needed columns
Refine the text data by removing stopwords, as well as placeholders like {{USERNAME}} and {{URL}}  and then  generate a new word cloud. This entails using Python's natural language processing tools like NLTK for handling stopwords and regular expressions for removing specific patterns.


```python

# Ensure stopwords are downloaded
nltk.download('stopwords')

# Assuming 'df' is your DataFrame and it contains columns 'text' and 'label'
df = df[['text', 'label']]  # Keep only relevant columns

# Function to clean text
def clean_text(text):
    # Replace placeholders with space
    text = re.sub(r'\{\{USERNAME\}\}', ' ', text)
    text = re.sub(r'\{\{URL\}\}', ' ', text)
    # Remove anything that is not a letter
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    # Convert text to lowercase
    text = text.lower()
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = text.split()
    text = ' '.join(word for word in words if word not in stop_words)
    return text

# Apply the cleaning function to the text
df['clean_text'] = df['text'].apply(clean_text)

# Collect all cleaned text into a single string for analysis
all_clean_text = ' '.join(df['clean_text'])

# Function to generate visualizations: Word Cloud and Bar Plot
def generate_new_visualizations(text):
    
    # Calculate the frequency of words and prepare for the bar plot
    word_freq = Counter(text.split())
    common_words = word_freq.most_common(20)
    words, counts = zip(*common_words)

    # Bar plot for the most frequent words
    fig_common_words = go.Figure(go.Bar(x=counts, y=words, orientation='h'))
    fig_common_words.update_layout(title='Most Common Words', xaxis_title='Word Count', yaxis_title='Words')
    fig_common_words.show()

    # Generate Word Cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    fig_wordcloud = px.imshow(wordcloud, title="Word Cloud")
    fig_wordcloud.update_xaxes(visible=False)
    fig_wordcloud.update_yaxes(visible=False)
    fig_wordcloud.show()


# Generate the visualizations
generate_new_visualizations(all_clean_text)
```

#### Now we see more common words that's clean and applicable.

```python
#
# Convert labels to categorical
y = to_categorical(df['label'])

# Split the data into training and testing sets
X_train_text, X_test_text, y_train, y_test = train_test_split(df['clean_text'], y, test_size=0.2, random_state=42)

# Function to oversample minority classes
def simple_oversample(X, y):
    max_size = max([np.sum(y == i) for i in range(len(np.unique(y)))])
    X_oversampled = []
    y_oversampled = []

    for class_index in np.unique(y):
        X_class = X[y == class_index]
        y_class = y[y == class_index]
        X_resampled, y_resampled = resample(X_class, y_class, replace=True, n_samples=max_size, random_state=42)
        X_oversampled.append(X_resampled)
        y_oversampled.append(y_resampled)

    return np.concatenate(X_oversampled), np.concatenate(y_oversampled)

# Apply oversampling to the training text data
X_train_text_oversample, y_train_oversample = simple_oversample(X_train_text.values.reshape(-1, 1), y_train.argmax(axis=1))
X_train_text_oversample = X_train_text_oversample.flatten()

# Modeling
# Train various machine learning and deep learning models.

# Apply CountVectorizer
count_vectorizer = CountVectorizer()
X_train_count = count_vectorizer.fit_transform(X_train_text)
X_test_count = count_vectorizer.transform(X_test_text)

# Apply TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train_text)
X_test_tfidf = tfidf_vectorizer.transform(X_test_text)

# Apply TF-IDF Vectorizer to oversampled data
X_train_tfidf_oversample = tfidf_vectorizer.fit_transform(X_train_text_oversample)
X_test_tfidf_oversample = tfidf_vectorizer.transform(X_test_text)
```

```python
#
# Convert labels to categorical
y = to_categorical(df['label'])

# Split the data into training and testing sets
X_train_text, X_test_text, y_train, y_test = train_test_split(df['clean_text'], y, test_size=0.2, random_state=42)

# Function to oversample minority classes
def simple_oversample(X, y):
    max_size = max([np.sum(y == i) for i in range(len(np.unique(y)))])
    X_oversampled = []
    y_oversampled = []

    for class_index in np.unique(y):
        X_class = X[y == class_index]
        y_class = y[y == class_index]
        X_resampled, y_resampled = resample(X_class, y_class, replace=True, n_samples=max_size, random_state=42)
        X_oversampled.append(X_resampled)
        y_oversampled.append(y_resampled)

    return np.concatenate(X_oversampled), np.concatenate(y_oversampled)

# Apply oversampling to the training text data
X_train_text_oversample, y_train_oversample = simple_oversample(X_train_text.values.reshape(-1, 1), y_train.argmax(axis=1))
X_train_text_oversample = X_train_text_oversample.flatten()

# Modeling
# Train various machine learning and deep learning models.

# Apply CountVectorizer
count_vectorizer = CountVectorizer()
X_train_count = count_vectorizer.fit_transform(X_train_text)
X_test_count = count_vectorizer.transform(X_test_text)

# Apply TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train_text)
X_test_tfidf = tfidf_vectorizer.transform(X_test_text)

# Apply TF-IDF Vectorizer to oversampled data
X_train_tfidf_oversample = tfidf_vectorizer.fit_transform(X_train_text_oversample)
X_test_tfidf_oversample = tfidf_vectorizer.transform(X_test_text)

# Train and evaluate Logistic Regression
def train_evaluate_lr(X_train, X_test, y_train, y_test):
    print('Training for Logistic Regression ')
    lr_model = LogisticRegression(max_iter=100)
    lr_model.fit(X_train, y_train.argmax(axis=1))
    y_pred = lr_model.predict(X_test)
    return y_pred

# Train and evaluate Naive Bayes
def train_evaluate_nb(X_train, X_test, y_train, y_test):
    print('Training for Naive Bayes')
    nb_model = MultinomialNB()
    nb_model.fit(X_train, y_train.argmax(axis=1))
    y_pred = nb_model.predict(X_test)
    return y_pred

# Train and evaluate Deep Learning model
def train_evaluate_dl(X_train, X_test, y_train, y_test):
    print('Training for Deep Learning model')
    model = Sequential([
        Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
        Dense(64, activation='relu'),
        Dense(y_train.shape[1], activation='softmax')
    ])

    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    history = model.fit(X_train, y_train, epochs=2, batch_size=64, validation_data=(X_test, y_test), verbose=2)
    
    y_pred_probs = model.predict(X_test)
    y_pred = np.argmax(y_pred_probs, axis=1)
    
    return y_pred
```

```python


# Evaluation
# Evaluate the performance of trained models.

# Function to print metrics
def print_metrics(model_name, y_test, y_pred, approach):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro')
    recall = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')
    auc = roc_auc_score(y_test, to_categorical(y_pred), multi_class='ovo')
    cm = confusion_matrix(y_test, y_pred)
    
    print(f"{model_name} ({approach}):")
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")
    print(f"AUC: {auc:.2f}")
#     print(f"Confusion Matrix:\n {cm}\n")
    print("Confusion Matrix:")
    sns.heatmap(cm, annot=True, fmt='d')
    plt.title(f'{model_name} {approach} Confusion Matrix')
    plt.show()
    
    return accuracy, precision, recall, f1

# Evaluate models with Count Vectorizer
y_pred_lr_count = train_evaluate_lr(X_train_count, X_test_count, y_train, y_test)
y_pred_nb_count = train_evaluate_nb(X_train_count, X_test_count, y_train, y_test)
y_pred_dl_count = train_evaluate_dl(X_train_count, X_test_count, y_train, y_test)

# Evaluate models with TF-IDF Vectorizer
y_pred_lr_tfidf = train_evaluate_lr(X_train_tfidf, X_test_tfidf, y_train, y_test)
y_pred_nb_tfidf = train_evaluate_nb(X_train_tfidf, X_test_tfidf, y_train, y_test)
y_pred_dl_tfidf = train_evaluate_dl(X_train_tfidf, X_test_tfidf, y_train, y_test)

# Evaluate models with TF-IDF Vectorizer on oversampled data
y_pred_lr_tfidf_oversample = train_evaluate_lr(X_train_tfidf_oversample, X_test_tfidf_oversample, to_categorical(y_train_oversample), y_test)
y_pred_nb_tfidf_oversample = train_evaluate_nb(X_train_tfidf_oversample, X_test_tfidf_oversample, to_categorical(y_train_oversample), y_test)
y_pred_dl_tfidf_oversample = train_evaluate_dl(X_train_tfidf_oversample, X_test_tfidf_oversample, to_categorical(y_train_oversample), y_test)

```

```python
# Print results for Logistic Regression
acc_lr_count, precision_lr_count, recall_lr_count, f1_lr_count = print_metrics("Logistic Regression", y_test.argmax(axis=1), y_pred_lr_count, "Count Vectorizer")
acc_lr_tfidf, precision_lr_tfidf, recall_lr_tfidf, f1_lr_tfidf = print_metrics("Logistic Regression", y_test.argmax(axis=1), y_pred_lr_tfidf, "TF-IDF Vectorizer")
acc_lr_tfidf_oversample, precision_lr_tfidf_oversample, recall_lr_tfidf_oversample, f1_lr_tfidf_oversample = print_metrics("Logistic Regression", y_test.argmax(axis=1), y_pred_lr_tfidf_oversample, "TF-IDF Oversample")

# Print results for Naive Bayes
acc_nb_count, precision_nb_count, recall_nb_count, f1_nb_count = print_metrics("Naive Bayes", y_test.argmax(axis=1), y_pred_nb_count, "Count Vectorizer")
acc_nb_tfidf, precision_nb_tfidf, recall_nb_tfidf, f1_nb_tfidf = print_metrics("Naive Bayes", y_test.argmax(axis=1), y_pred_nb_tfidf, "TF-IDF Vectorizer")
acc_nb_tfidf_oversample, precision_nb_tfidf_oversample, recall_nb_tfidf_oversample, f1_nb_tfidf_oversample = print_metrics("Naive Bayes", y_test.argmax(axis=1), y_pred_nb_tfidf_oversample, "TF-IDF Oversample")

# Print results for Deep Learning
acc_dl_count, precision_dl_count, recall_dl_count, f1_dl_count = print_metrics("Deep Learning", y_test.argmax(axis=1), y_pred_dl_count, "Count Vectorizer")
acc_dl_tfidf, precision_dl_tfidf, recall_dl_tfidf, f1_dl_tfidf = print_metrics("Deep Learning", y_test.argmax(axis=1), y_pred_dl_tfidf, "TF-IDF Vectorizer")
acc_dl_tfidf_oversample, precision_dl_tfidf_oversample, recall_dl_tfidf_oversample, f1_dl_tfidf_oversample = print_metrics("Deep Learning", y_test.argmax(axis=1), y_pred_dl_tfidf_oversample, "TF-IDF Oversample")

# Display results in tabular format
# Logistic Regression Results
lr_results = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1 Score"],
    "Before Oversampling": [acc_lr_tfidf, precision_lr_tfidf, recall_lr_tfidf, f1_lr_tfidf],
    "After Oversampling": [acc_lr_tfidf_oversample, precision_lr_tfidf_oversample, recall_lr_tfidf_oversample, f1_lr_tfidf_oversample]
})

# Naive Bayes Results
nb_results = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1 Score"],
    "Before Oversampling": [acc_nb_tfidf, precision_nb_tfidf, recall_nb_tfidf, f1_nb_tfidf],
    "After Oversampling": [acc_nb_tfidf_oversample, precision_nb_tfidf_oversample, recall_nb_tfidf_oversample, f1_nb_tfidf_oversample]
})

# Deep Learning Results
dl_results = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1 Score"],
    "Before Oversampling": [acc_dl_tfidf, precision_dl_tfidf, recall_dl_tfidf, f1_dl_tfidf],
    "After Oversampling": [acc_dl_tfidf_oversample, precision_dl_tfidf_oversample, recall_dl_tfidf_oversample, f1_dl_tfidf_oversample]
})

print("Logistic Regression:")
print(lr_results)
print("\nNaive Bayes:")
print(nb_results)
print("\nDeep Learning:")
print(dl_results)

# Display results in tabular format
# Logistic Regression Results
lr_results = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1 Score"],
    "Before Oversampling": [acc_lr_tfidf, precision_lr_tfidf, recall_lr_tfidf, f1_lr_tfidf],
    "After Oversampling": [acc_lr_tfidf_oversample, precision_lr_tfidf_oversample, recall_lr_tfidf_oversample, f1_lr_tfidf_oversample]
})

# Naive Bayes Results
nb_results = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1 Score"],
    "Before Oversampling": [acc_nb_tfidf, precision_nb_tfidf, recall_nb_tfidf, f1_nb_tfidf],
    "After Oversampling": [acc_nb_tfidf_oversample, precision_nb_tfidf_oversample, recall_nb_tfidf_oversample, f1_nb_tfidf_oversample]
})

# Deep Learning Results
dl_results = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1 Score"],
    "Before Oversampling": [acc_dl_tfidf, precision_dl_tfidf, recall_dl_tfidf, f1_dl_tfidf],
    "After Oversampling": [acc_dl_tfidf_oversample, precision_dl_tfidf_oversample, recall_dl_tfidf_oversample, f1_dl_tfidf_oversample]
})

print("Logistic Regression:")
print(lr_results)
print("\nNaive Bayes:")
print(nb_results)
print("\nDeep Learning:")
print(dl_results)
```

## Evaluation and Discussion of Result
In this section, we will be evaluating the result of the three predictive models, which is logistic regression, Naive Bayes and Deep Learning.
### Evaluation
The first step in the evaluation was the selection of three models, namely: Logistic Regression, Naive Bayes, and Deep Learning. Each of the three models was trained and evaluated using three different preprocessing and vectorization techniques: Count Vectorizer, TF-IDF Vectorizer, and Oversampling with TF-IDF Vectorizer.  
The goal of using ​​Count Vectorizer and TF-IDF Vectorizer is to see how the vectorization techniques affect performance, while the goal of the Oversampling with TF-IDF Vectorizer is to address the class imbalance problem and see how the model performs without it.
The performance metrics used include accuracy, precision, recall, and F1 score for each combination of model and vectorization technique.

### Interpretation of Results:
#### Result of Logistic Regression
* Performance: For Logistic Regression, we can observe that oversampling with TF-IDF vectorization improved performance across most of the performance metrics. This indicates the importance of addressing the issue of class imbalance.
* Weaknesses: TF-IDF had the lowest performance score in the various metrics used 
* Strengths: Overall, the count vectorizer gave a good performance score for all metrics despite the  imbalance of the dataset
#### Result for Naive Bayes
* Performance: Naive Bayes, despite poor performance with TF-IDF vectorization, showed great improvement in metrics such as recall and F1 score after oversampling, indicating that class imbalances can enhance probabilistic models.
* Weaknesses: Poor performance with TF-IDF vectorization,
* Strengths: Significant improvement after Oversampling.
#### Result for Deep Learning
* Performance: ​Showed strong performance across different vectorization techniques. While it showed slight improvements in oversampling, the improvements were not as significant in comparison to the other models. This suggests that deep learning models inherently handle class imbalances better or that they benefit less from oversampling techniques
* Weaknesses: Low recall and F1 Score value with TF-IDF
* Strengths: Performed better than the other two models across the board and appears capable of dealing with an imbalanced dataset.
#### Overall Performance
From the results, it's evident that oversampling with TF-IDF has improved the performance of all models across the various metrics, especially for Logistic Regression and Naive Bayes. 
Also, the Deep Learning Model consistently outperformed other models across the different vectorization techniques but showed little improvement with an oversampled dataset. Lastly for this experiment, the Count Vectorizer outperformed the TF-IDF Vectorizer across the performance metrics.
